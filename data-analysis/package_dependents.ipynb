```python
package_names = load_package_names()

temp = package_names[['gh_url_api']].copy()

def extract_repo_name(row):
    try:
        return row['gh_url_api'].split('https://api.github.com/repos/')[1]
    except Exception:
        return pd.NA
temp['repo_name'] = package_names.apply(extract_repo_name, axis=1)


def extract_name(row):
    try:
        return row['repo_name'].split('/')[1]
    except Exception:
        return pd.NA

temp['name'] = temp.apply(extract_name, axis=1)
temp['name'].value_counts()
```
